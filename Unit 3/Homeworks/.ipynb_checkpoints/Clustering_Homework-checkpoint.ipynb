```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./datasets/clustering_data.csv", header=None)
data
```

```python
plt.scatter(data[0],data[1])
```

```python
distorsions = []
for i in range(2,20):
    model = KMeans(n_clusters=i)
    model.fit(data)
    distorsions.append(model.inertia_)

plt.figure(figsize=(15,5))
plt.plot(range(2,20),distorsions)
plt.grid(True)
```

```python
# The value that represents the elbow is 6
model = KMeans(n_clusters=6, init='random',random_state=1)
model.fit(data)
y_pred = model.predict(data)
plt.scatter(data[0],data[1],c=y_pred)
```
