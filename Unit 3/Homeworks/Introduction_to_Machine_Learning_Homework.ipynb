```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Reading the data

```python
data = pd.read_csv("datasets/food-allergy-analysis-Zenodo.csv")
```

```python
data.head()
```

```python
print(len(data))
data.isna().sum()
```

```python
data.insert(len(data.columns), 'Got_Food_alg', np.ones((len(data),), dtype=bool))
for i in range(len(data)):
    data.at[i,'Got_Food_alg'] = data.iloc[i, 9:-1].any()

data
```

```python
alg_cols = data.columns[9:-1]
data.drop(alg_cols, axis=1, inplace=True)
data.drop(["SUBJECT_ID"], axis=1, inplace=True)
data["Got_Food_alg"] = data["Got_Food_alg"].astype(int)
data["ATOPIC_MARCH_COHORT"] = data["ATOPIC_MARCH_COHORT"].astype(int)
for i in range(len(data)):
    data.at[i,"GENDER_FACTOR"] = int(data.at[i,"GENDER_FACTOR"][1])
    data.at[i,"RACE_FACTOR"] = int(data.at[i,"RACE_FACTOR"][1])
    data.at[i,"ETHNICITY_FACTOR"] = int(data.at[i,"ETHNICITY_FACTOR"][1])
    data.at[i,"PAYER_FACTOR"] = int(data.at[i,"PAYER_FACTOR"][1])

data["GENDER_FACTOR"] = data["GENDER_FACTOR"].astype(int)
data["RACE_FACTOR"] = data["RACE_FACTOR"].astype(int)
data["ETHNICITY_FACTOR"] = data["ETHNICITY_FACTOR"].astype(int)
data["PAYER_FACTOR"] = data["PAYER_FACTOR"].astype(int)
data
```

```python
data.info()
```

# The logistic regression model

```python
from sklearn.model_selection import train_test_split

X = data.drop(['Got_Food_alg'],axis=1)
y = data['Got_Food_alg']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
score = model.score(X_test,y_test)

score
```
